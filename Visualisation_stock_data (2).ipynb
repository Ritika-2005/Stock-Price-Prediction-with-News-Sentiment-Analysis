```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# WIPRO Stock Data

# Load data
stock_data_wipro = pd.read_csv('data/WIPRO_data.csv')

stock_data_wipro['Date'] = pd.to_datetime(stock_data_wipro['Date'], errors='coerce')
```

```python
# Visualisation of stock data

# Stock Return as the percentage change in 'adj_close'
stock_data_wipro['Stock Return'] = stock_data_wipro['adj_close'].pct_change() * 100

# Set the date as index 
stock_data_wipro.set_index('Date', inplace=True)

# Create subplots
fig, axes = plt.subplots(4, 1, figsize=(14, 12), sharex=True)

# Plot Stock Return
sns.lineplot(ax=axes[0], x=stock_data_wipro.index, y=stock_data_wipro['Stock Return'], color='blue')
axes[0].set_ylabel('Stock Return (%)')
axes[0].set_title('WIPRO Stock Data')

# Plot Closing Price
sns.lineplot(ax=axes[1], x=stock_data_wipro.index, y=stock_data_wipro['close'], color='green')
axes[1].set_ylabel('Closing Price')

# Plot Shares Traded (Volume)
sns.lineplot(ax=axes[2], x=stock_data_wipro.index, y=stock_data_wipro['volume'], color='red')
axes[2].set_ylabel('Shares Traded')

# Plot Adjusted Close
sns.lineplot(ax=axes[3], x=stock_data_wipro.index, y=stock_data_wipro['adj_close'], color='purple')
axes[3].set_ylabel('Adjusted Close')
axes[3].set_xlabel('Year')

plt.show()

stock_data_wipro.reset_index(inplace=True)
```
